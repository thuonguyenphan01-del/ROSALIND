```python
from Bio import SeqIO
import numpy as np
seq_name, seq_string = [], []
with open (r"C:\Users\ASUS\Downloads\rosalind_cons.txt",'r') as fa:
    for seq_record  in SeqIO.parse(fa,'fasta'):
        seq_name.append((seq_record.name))
        seq_string.append(str(seq_record.seq))#chuyển dạng string

seq_len = len(seq_string)
str_len = len(seq_string[0])
nu = ["A", "C", "G", "T"] #tạo list nu
c_string= "" # gắn giá trị string
matrix = np.zeros(shape=(4, str_len), dtype=int)# tạo matrix
for c in range(str_len):
     convert_co= [s[c] for s in seq_string] # tạo cột nu
     max_appeared= max(convert_co,key=convert_co.count) #nu xuất hiện nhiều lần nhất theo cột
     c_string+= max_appeared
     for n in range(len(nu)):
         matrix[n][c]= convert_co.count(nu[n])
print(c_string) 
for i in range(len(nu)):
    print(f"{nu[i]}: {" ".join([str(j) for j in matrix[i]])}")


```
