```python
def rab(n, m):
    tuoi=[0]*m # tao mảng tuổi tuoi[i] với i từ 0 -> m-1 m là tuổi thọ tối đa
    tuoi[0]=1 # tháng đầu tiên có 1 con thỏ
    for i in range(1,n):
        new= sum(tuoi[1:]) #tính tổng các con thỏ đã đủ tuổi sinh nở >=1. Mỗi cặp thỏ sẽ cho ra 1 cặp thỏ vậy số lượng thỏ được sinh ra bằng tổng số lượng thỏ trưởng thành
        tuoi= [new] + tuoi[:-1] # lớp thỏ già nhất đã chết nên bỏ số thỏ ở hàng cuối cùng
    return sum(tuoi)
print(rab(90,16))
        
```
