```python
from Bio import SeqIO
seq, seq_id =[],[]
with open (r"C:\Users\ASUS\Downloads\rosalind_grph (1).txt","r") as f:
    for record in SeqIO.parse(f, "fasta"):
        seq_id.append((record.name))
        seq.append(str(record.seq))
n=3
for i in range(len(seq_id)):#chạy vòng lặp từ gen 0
    for j in range(len(seq_id)):#lồng thêm vòng lặp chạy từ g gen 0
        if i!=j: # không trùng đoạn gen
            if seq[i][-n:]==seq[j][:n]:
                print(seq_id[i],seq_id[j])


```
