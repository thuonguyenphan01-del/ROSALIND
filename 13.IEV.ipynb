Vậy xác xuất sinh con mang kiểu hình trội:
P = 2*(a+b+c+0.75*d+0.5*e)

```python
def dominant_alle(a,b,c,d,e,f):
    P = 2*(a+b+c+0.75*d+0.5*e+0*f)
    return P

with open(r"C:\Users\ASUS\Downloads\rosalind_iev.txt","r") as f:
    line=f.readline()
ln= list(line.split(" "))
lst= [int(x.strip()) for x in ln]
print(dominant_alle(*lst))


```
