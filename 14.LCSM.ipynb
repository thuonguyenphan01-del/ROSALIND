```python
from Bio import SeqIO
seq_id,seq_string=[],[]
with open (r"C:\Users\ASUS\Downloads\rosalind_lcsm.txt","r") as f:
    for record in SeqIO.parse(f, 'fasta'):
        seq_id.append(record.name)
        seq_string.append(str(record.seq))
def longest_common_substring(seqs):
    shortest_seq=min(seqs,key=len)
    number_string_shortest=len(shortest_seq)
    for length in range(number_string_shortest,0,-1):#dò từ đoạn có kích thước lớn đến kich sthuwocs nhỏ
        for sta_point in range(number_string_shortest-length+1):#có bao nhiêu chuổi có kích thức n trong chuỗi trên
            match_seq= shortest_seq[sta_point:sta_point+length]
            if all(match_seq in s for s in seq_string):
                return(match_seq)# Nếu tìm thấy chuỗi chung
    return("")# Nếu không trả về ""
print(longest_common_substring(seq_string))


        
```
