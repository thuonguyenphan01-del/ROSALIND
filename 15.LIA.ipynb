- Xác suất một cá thể có kiểu gen Aa Bb
Với locus A: từ Aa × Aa → xác suất con là Aa = 1/2.
Với locus B: từ Bb × Bb → xác suất con là Bb = 1/2.
Hai locus độc lập → xác suất con là Aa Bb = 1/2* 1/2=1/4.
- Số cá thể trong thế hệ k
Mỗi cá thể sinh 2 con → thế hệ k có 2^k cá thể.

```python
from math import comb

def prob_at_least_N_AaBb(k, N):
    n = 2 ** k
    p = 1/4
    s = 0.0
    for i in range(N):
        s += comb(n, i) * (p ** i) * ((1 - p) ** (n - i))
    return 1 - s

print(f"{prob_at_least_N_AaBb(6, 17):.3f}") 
```
