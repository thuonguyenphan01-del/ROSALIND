```python
def mrna(protein):
    codons = {'F': ['UUU', 'UUC'],
              'L': ['UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG'],
              'S': ['UCU', 'UCC', 'UCA', 'UCG', 'AGU', 'AGC'],
              'Y': ['UAU', 'UAC'],
              '*': ['UAA', 'UAG', 'UGA'],
              'C': ['UGU', 'UGC'],
              'W': ['UGG'],
              'P': ['CCU', 'CCC', 'CCA', 'CCG'],
              'H': ['CAU', 'CAC'],
              'Q': ['CAA', 'CAG'],
              'R': ['CGU', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG'],
              'V': ['GUU', 'GUC', 'GUA', 'GUG'],
              'A': ['GCU', 'GCC', 'GCA', 'GCG'],
              'D': ['GAU', 'GAC'],
              'E': ['GAA', 'GAG'],
              'G': ['GGU', 'GGC', 'GGA', 'GGG'],
              'I': ['AUU', 'AUC', 'AUA'],
              'M': ['AUG'],
              'T': ['ACU', 'ACC', 'ACA', 'ACG'],
              'N': ['AAU', 'AAC'],
              'K': ['AAA', 'AAG']}
    number = 1
    for i in protein:
        number = number * len(codons[i])
    number = number*len(codons["*"])
    return number % 1000000

with open(r"C:\Users\ASUS\Downloads\rosalind_mrna.txt", 'r') as f:
    protein = f.readline().strip()
    number = mrna(protein)
    print(number)
```

Với mỗi amino acid trong chuỗi protein, ta nhân số cách mã hóa của nó.
Ví dụ protein "MA":
'M' có 1 codon.
'A' có 4 codons.
Tổng số cách = 1* 4=4.
Cuối cùng, mọi chuỗi mRNA phải kết thúc bằng một stop codon.
Có 3 cách chọn stop codon.
Vì vậy ta nhân thêm 3.
