```python
import pandas as pd
from Bio import SeqIO
with open (r"C:\Users\ASUS\Downloads\rosalind_orf (1).txt", "r") as f:
    data= SeqIO.to_dict(SeqIO.parse(f,"fasta"))
dna= str(data['Rosalind_0900'].seq)
format={'A':"U","T":"A","G":"C","C":"G"}
reverse_rna=("".join(format[dna[i]] for i in range(len(dna))))
reverse_rna=reverse_rna[::-1]
rna=dna.replace("T","U")
answer=[]
def mahoa(rna):
    gene=rna.split("AUG")
    for i in range(len(gene)):
        pro="M" + "".join(codon[gene[i][j:j+3]] if (j < len(gene[i])-2) else "" for j in range(len(gene[i])))
        pro=pro.split("Stop")
        answer.append(pro)
mahoa(reverse_rna)
mahoa(rna)
final=[]
for i in range(len(answer)):
    for j in range(len(answer[i])):
        if answer[i][j][0]=="M":
            final.append(answer[i][j])
final=pd.DataFrame(final)
p=final[0].unique()
for i in range(len(p)):
    print(p[i])
```
