sorted(alphabet): sắp xếp alphabet để kết quả đúng thứ tự từ điển.
itertools.product(alphabet, repeat=n): sinh tất cả các tổ hợp có lặp độ dài n.
Ví dụ: với alphabet = ['A','B'], repeat=2 → (A,A), (A,B), (B,A), (B,B).
''.join(p): ghép tuple thành chuỗi.


```python
import itertools

def lex_strings(alphabet, n):
    alphabet = sorted(alphabet)  # đảm bảo theo thứ tự chuẩn
    result = [''.join(p) for p in itertools.product(alphabet, repeat=n)]
    return result

with open(r"C:\Users\ASUS\Downloads\rosalind_lexf.txt", 'r') as f:
    string = f.readline().split()
    n = int(f.readline().strip())
strings = lex_strings(string, n)
print("\n".join(strings))
```
