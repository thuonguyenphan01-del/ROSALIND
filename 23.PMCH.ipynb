Input: một chuỗi RNA (FASTA format).
Output: số lượng perfect matchings của các base trong chuỗi RNA.
Quy tắc bắt cặp:
A ↔ U
C ↔ G
Nếu chuỗi RNA có số lượng A bằng số lượng U và số lượng C bằng số lượng G, thì số lượng perfect matchings = a!* c!.


```python
from Bio import SeqIO
def factorial(n):
    f = 1
    for i in range(n):
        f *= (i+1)#giai thừa
    return f

def pmch(s):
    a = s.count('A')
    c = s.count('C')
    print(a, c)
    return factorial(a) * factorial(c)
if __name__ == "__main__":
    seq_name, seq_string = [], []
    with open (r"C:\Users\ASUS\Downloads\rosalind_pmch.txt",'r') as fa:
        for seq_record  in SeqIO.parse(fa,'fasta'):
            seq_name.append(str(seq_record.name))
        seq_string.append(str(seq_record.seq))
print(pmch(seq_string[0]))
```
