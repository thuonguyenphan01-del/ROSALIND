Input:
Một chuỗi DNA s.
Một danh sách các giá trị GC‑content (tỉ lệ G+C trong chuỗi).
Output:
Với mỗi giá trị GC‑content p, in ra logarithm cơ số 10 của xác suất chuỗi s xuất hiện nếu được sinh ngẫu nhiên với GC‑content đó.
Xác suất để một ký tự A hoặc T xuất hiện = (1-i)/2
Xác suất để một ký tự C hoặc G xuất hiện = i/2



```python
import math

with open(r"C:\Users\ASUS\Downloads\rosalind_prob.txt", "r") as f:
    s = f.readline().strip()
    A = map(float, f.readline().strip().split(" "))

a = s.count('A') + s.count('T')
c = s.count('C') + s.count('G')
for i in A:
    p = a * math.log((1-i)/2, 10) + c * math.log(i/2, 10)
    print(p, end=" ")
```
