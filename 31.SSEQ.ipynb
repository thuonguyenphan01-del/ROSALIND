Input: hai chuỗi DNA s và t.
Output: vị trí (theo chỉ số 1-based) của các ký tự trong t khi chúng xuất hiện như một subsequence trong s.
Nói cách khác: tìm cách "ghép" chuỗi t vào s theo đúng thứ tự, rồi in ra vị trí từng ký tự.
- vì đề Rosalind yêu cầu chỉ số 1-based (bắt đầu từ 1), trong khi Python trả về chỉ số 0-based nên phải +1

```python
#id:SSEQ
from Bio import SeqIO
def find_spliced_motif(s, t):
    pos = []
    start = 0
    for ch in t:
        start = s.index(ch, start) + 1   # tìm từ vị trí start trở đi
        pos.append(start)
    print(" ".join(map(str, pos)))

if __name__ == "__main__":
    seq_name, seq_string = [], []
    with open (r"C:\Users\ASUS\Downloads\rosalind_sseq.txt",'r') as fa:
        for seq_record  in SeqIO.parse(fa,'fasta'):
            seq_name.append(str(seq_record.name))
            seq_string.append(str(seq_record.seq))

    s, t = seq_string[0], seq_string[1]
    find_spliced_motif(s,t)
```
