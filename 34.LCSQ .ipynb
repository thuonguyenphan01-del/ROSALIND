Input: 2 chuỗi DNA.
Output: chuỗi con chung dài nhất (Longest Common Subsequence – LCS).

```python
from Bio import SeqIO

def lcs(s, t):
    # Khởi tạo ma trận DP bằng list comprehension ma trận 2 chiều len(t)+1 x len(s)+1
    dp = [[0]*(len(t)+1) for _ in range(len(s)+1)]

    # Điền ma trận
    for i in range(1, len(s)+1):
        for j in range(1, len(t)+1):
            dp[i][j] = dp[i-1][j-1]+1 if s[i-1]==t[j-1] else max(dp[i-1][j], dp[i][j-1])

    # Truy vết để dựng chuỗi LCS
    i, j, res = len(s), len(t), []
    while i>0 and j>0:
        if s[i-1]==t[j-1]:
            res.append(s[i-1])
            i, j = i-1, j-1
        elif dp[i-1][j] >= dp[i][j-1]:
            i -= 1
        else:
            j -= 1
    return ''.join(reversed(res))

with open(r"C:\Users\ASUS\Downloads\rosalind_lcsq (1).txt") as f:
    seqs = [str(rec.seq) for rec in SeqIO.parse(f,"fasta")]

print(lcs(seqs[0], seqs[1]))
```
