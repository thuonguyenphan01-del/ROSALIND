```python
def shortest_common_supersequence(s, t):
    # Tìm LCS bằng DP
    m, n = len(s), len(t)
    dp = [[""] * (n+1) for _ in range(m+1)]
    
    for i in range(m):
        for j in range(n):
            if s[i] == t[j]:
                dp[i+1][j+1] = dp[i][j] + s[i]
            else:
                dp[i+1][j+1] = max(dp[i][j+1], dp[i+1][j], key=len)
    
    lcs = dp[m][n]
    
    # Xây dựng SCS từ LCS
    i = j = 0
    scs = []
    for c in lcs:
        while s[i] != c:
            scs.append(s[i])
            i += 1
        while t[j] != c:
            scs.append(t[j])
            j += 1
        scs.append(c)
        i += 1
        j += 1
    
    scs.extend(s[i:])
    scs.extend(t[j:])
    
    return "".join(scs)
with open(r"C:\Users\ASUS\Downloads\rosalind_scsp.txt", "r") as f:
    s = f.readline().strip()
    t = f.readline().strip()

print(shortest_common_supersequence(s, t))
```
