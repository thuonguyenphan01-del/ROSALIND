```python
#REVP
from Bio.Seq import Seq
from Bio import SeqIO

def find_reverse_palindromes(seq):
    results = []
    n = len(seq)
    for i in range(n):
        for length in range(4, 13):  # từ 4 đến 12
            if i + length <= n:
                sub = seq[i:i+length]
                if sub == str(Seq(sub).reverse_complement()):
                    results.append((i+1, length))  # 1-based index
    return results

fasta_file = r"C:\Users\ASUS\Downloads\rosalind_revp.txt" 
record = SeqIO.read(fasta_file, "fasta")
dna_seq = str(record.seq)

results = find_reverse_palindromes(dna_seq)
for pos, length in results:
    print(pos, length)
```
