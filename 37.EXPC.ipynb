- Cho GC-content = p.
- Khi đó:
- Xác suất của ký tự G hoặc C = p/2.
- Xác suất của ký tự A hoặc T = (1-p)/2.
- m=2.
- Với p=0.25:
- P(A) = (1-0.25)/2 = 0.375
- P(G) = 0.25/2 = 0.125
- P(s = "AG") = 0.375 × 0.125 = 0.046875
- Expected = (10-2+1) × 0.046875 = 9 × 0.046875 = 0.422

```python
#id EXPC
def expected_counts(n, s, gc_array):
    m = len(s)
    results = []
    for gc in gc_array:
        prob = 1.0
        for ch in s:
            if ch in "GC":
                prob *= gc / 2
            else:
                prob *= (1 - gc) / 2
        expected = (n - m + 1) * prob
        results.append(round(expected, 3))
    return results

with open(r"C:\Users\ASUS\Downloads\rosalind_eval.txt", "r") as f:
    n = int(f.readline().strip())
    s = f.readline().strip()
    gc_array = list(map(float, f.readline().split()))

print(" ".join(map(str, expected_counts(n, s, gc_array))))
```
