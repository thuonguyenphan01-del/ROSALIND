```python
#id SIGN
from itertools import permutations, product

def signed_permutations(n):
    nums = list(range(1, n+1))
    perms = permutations(nums)
    results = []
    for p in perms:
        for signs in product([-1, 1], repeat=n):
            signed = [a * b for a, b in zip(p, signs)]
            results.append(signed)
    return results

with open(r"C:\Users\ASUS\Downloads\rosalind_sign.txt", "r") as f:
    n = int(f.read().strip())

results = signed_permutations(n)
print(len(results))
for r in results:
    print(" ".join(map(str, r)))
```
