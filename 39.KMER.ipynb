```python
#KMER
from Bio import SeqIO
from itertools import product

def kmer_composition(seq, k=4):
    # sinh tất cả k-mers theo lexicographic order
    kmers = ["".join(p) for p in product("ACGT", repeat=k)]
    counts = {kmer: 0 for kmer in kmers}
    
    # đếm số lần xuất hiện
    for i in range(len(seq) - k + 1):
        sub = seq[i:i+k]
        counts[sub] += 1
    
    # trả về mảng theo đúng thứ tự
    return [counts[kmer] for kmer in kmers]

fasta_file = r"C:\Users\ASUS\Downloads\rosalind_kmer.txt"
record = SeqIO.read(fasta_file, "fasta")
dna_seq = str(record.seq)

result = kmer_composition(dna_seq, 4)
print(" ".join(map(str, result)))
```
