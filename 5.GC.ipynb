```python
from Bio import SeqIO

max_seq_id = ""
max_gc = 0
with open (r"C:\Users\ASUS\Downloads\rosalind_gc (3).txt","r") as fa:
    for record in SeqIO.parse(fa, "fasta"):
        sequence = str(record.seq) # convert sequence to string
        sequence_id = record.id
        gc_content = (sequence.count("C") + sequence.count("G")) / len(sequence) * 100
        if gc_content > max_gc:
            max_seq_id = sequence_id
            max_gc = gc_content
print(f"{max_seq_id} {max_gc:5f}%")
```
