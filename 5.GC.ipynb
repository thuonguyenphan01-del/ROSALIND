```python
from Bio import SeqIO

max_seq_id = ""
max_gc = 0
with open (r"C:\Users\ASUS\Downloads\rosalind_gc (3).txt","r") as fa:
    for record in SeqIO.parse(fa, "fasta"):
        sequence = str(record.seq) 
        sequence_id = record.id
        gc_content = (sequence.count("C") + sequence.count("G")) / len(sequence) * 100
        if gc_content > max_gc:
            max_seq_id = sequence_id
            max_gc = gc_content
print(f"{max_seq_id} {max_gc:5f}%")
```

```python
from Bio import SeqIO
import pandas as pd
seq_id,seq=[],[]
with open (r"C:\Users\ASUS\Downloads\rosalind_gc (3).txt","r") as fa:
    for record in SeqIO.parse(fa, "fasta"):
        seq.append(str(record.seq)) 
        seq_id.append(record.id)
table=pd.DataFrame(seq_id)
table[1]=seq
per_GC=[]
for i in range(len(table[1])):
    G=table[1][i].count("G")
    C=table[1][i].count("C")
    m=len(table[1][i])
    per=round((G+C)*100/m,5)
    per_GC.append(per)
table[2]=per_GC
a=table.nlargest(1,2).iloc[0]
print(f"{a[0]} {a[2]}%")
```
