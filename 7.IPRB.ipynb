Khả năng sinh đồng lặn khi bố mẹ đều là đồng hợp lặn:
âa x aa: (n/t) * (n-1)/(t-1).
khả năng sinh ra đồng hợp khi bô mẹ đều di hợp
Aa x Aa: (m/t)*(m-1/t-1)*1/4.
Khả năng sinh ra đồng dị hợp khi bố mẹ 1 người đồng hợp 1 người dị hợp
aa x Aa: ((m/t)*(n/t-1) + (n/t)*(m/t-1))*1/2.
Khả năng sinhh ra đồng hợp lặn:
cộng hết lại

```python
def do(k,m,n):#k: homozygous dominant (AA) m: heterozygous (Aa) n: homozygous recessive (aa)
    t=k+m+n
    rr = n * (n - 1) / (t * (t - 1))   #aaxaa       
    rh = (m * n) / (t * (t - 1))      # Aa xaa        
    hh = 0.25 * (m * (m - 1) / (t * (t - 1))) # Aa x Aa
    d = rr + rh + hh
    return 1- d
print(f"{do(26,29,19):.5f}")
```
